```python
import pandas as pd
import numpy as np
import pandas as pd #kevin

df = pd.read_feather('data\echr.feather')
```

```python
pd.set_option('display.large_repr', 'truncate')
pd.set_option('display.max_columns', 0)
df.head()
```

```python
df.info()
# problems with types
```

```python
for column in df.columns:
    print(column)
    print(df[column].unique())
    print('__________________________')
```

```python
# Replace empty strings with NAN
df = df.apply(lambda x: x.str.strip()).replace('', np.nan)
df.info()
```

```python
# Convert to numeric
num_col = ('ECHRRanking', 'Rank', 'importance', 'languagenumber', 'originatingbody', 'sharepointid')
for col in num_col:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Convert to date
date_col = ('decisiondate', 'introductiondate', 'judgementdate', 'kpdate', 'kpdateAsText', 'referencedate', 'reportdate', 'resolutiondate')

# Convert to boolean
bool_col = ('isplaceholder', 'separateopinion')

# Convert to category
cat_col = ('application', 'doctype', 'doctypebranch', 'languageisocode', 'meetingnumber', 'typedescription')

for type_col, col_type in ((date_col, 'datetime64'), (bool_col, 'bool'), (cat_col, 'category')):
    for col in type_col:
        df[col] = df[col].astype(col_type)
```
